#Getting Address of a Variable

Use id function for it.

The id is the object's memory address, and will be different for each time you run the program.

The id() function returns a unique id for the specified object.

All objects in Python has its own unique id.

The id is assigned to the object when it is created.

```python
x = 5
y = 5
z = 10
print(x)
print(y)
print(id(x))
print(id(y))
print(id(z))
```

Comparison operators: ==, !=, >, <, >=, <=

Logical and: and

Logical or: or

#Assignment Operators

**Assignment operators** are used to assign values to variables.
The most basic operator is =, but there are others that perform operations while assigning.

*   = : Assigns the value on the right to the variable on the left.
*   += : Adds the right-hand value to the left-hand variable and assigns the result to the variable.
*   -= : Subtracts the right-hand value from the left-hand variable and assigns the result to the variable.
*   *= : Multiplies the left-hand variable by the right-hand value and assigns the result.
*   /= : Divides the left-hand variable by the right-hand value and assigns the result.










```python
x = 7
#x += 2     #Equivalent to x = x+2
#x -= 3      # Equivalent to x = x - 3
#x *= 4       # Equivalent to x = 5*5
#x /= 3        # Equivalent to x = x/3
#x %= 3
#x //= 3
#x **= 3
print(x)
```

#Membership Operators

**Membership operators** are used to check if a value exists in a sequence (like lists, strings, or tuples).
*   `in` Returns "True" if a sequence with the specified value is present in the object
*   `not in` Inline code Returns "True" if a sequence with the specified value is not present in the object


```python
var1 = "Python for Mathematicians"
print("y" in var1)


var2 = ["orange", 5, "3.5"]
print("orange" not in var2)

print("O" in "iCodeGuru")
```

#Identity Operators

**Identity operators** are used to check whether two variables refer to the same object in memory.

*   `is` : Returns True if both operands refer to the same object.
*   `is not` : Returns True if both operands do not refer to the same object.



```python
a = 10.1
b = 10.1

print(a is b)
print(id(a))
print(id(b))

#In the case of floating-point numbers, Python may or may not create separate objects for a and b,
# depending on how the memory is managed and Python’s internal optimizations.
```

```python
x = 10
y = 10
print(x is y)  # True because both refer to the same integer object
#Immutable objects (like integers, strings)
#can sometimes share memory if they hold the same value (an optimization Python makes).
```

```python
a = 10
b = "10"
print(a is b)
print(id(a))
print(id(b))
```

```python
a = [1, 2, 3]
b = a  # b refers to the same list object as a
c = [1, 2, 3]  # c is a different list object with the same content

print(a is b)      # True, because both refer to the same object
print(a is c)      # False, because a and c refer to different objects
print(a is not c)  # True, because a and c are not the same object
print(id(a))
print(id(b))
```
