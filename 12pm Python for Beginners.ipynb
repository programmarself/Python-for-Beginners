To add items from another set into the current set, use the `update()` method.

```python
set1 = {"Lahore", "Faisalabad", 5, True}
# print(dir(set1))
# print(help(set1))
set1.add("Peshawar")
print(set1)
```

```python
# update() is useful when you want to merge different collections of unique items.
'''This can be any iterable object, like a list, tuple, set, string,
or dictionary (only keys are added from a dictionary).'''

set1 = {1, 2, 3}

# Adding multiple elements from a list
set1.update([3, 4, 5])
print(set1)

# Adding elements from a tuple
set1.update((6, 7))
print(set1)

# Adding elements from another set
set1.update({8, 9})
print(set1)

set1.update("Pakistan")
print(set1)
```

```python
set1 = {"Lahore", "Faisalabad", 5, True}
set2 = {5, 0, False}
set1.update(set2)
print(set1)
```

#**Some Set Methods**

* `intersection_update()`
* `intersection()`
* `symmetric_difference()`
* `difference()`

```python
#The intersection_update() method will keep only the items that are present in both sets.

set1 = { 5, True,"python",7}
set2 = {5,6, "python"}
set2.intersection_update(set1)
print(set2)
print(set1)
```

```python
#The intersection() method will return a new set, that only contains the items that are present in both sets.

set1 = {"Lahore", "Faisalabad", 5,}
set2 = {5, 4,"Multan"}
x = set1.intersection(set2)
print(x)
print(set2)
```

```python
#difference method Returns a new set containing the difference between two or more sets
#Finds elements that are in the first set but not in the second set.

set1 = {"Lahore", "Faisalabad", 5,6}
set2 = {5,0,7,"Multan"}
x = set1.difference(set2)    # or set1 - set2
print(x)
```

```python
# Returns a new set containing elements that are unique to each set.
# Finds elements that are in either of the sets but not in both.

set1 = {"Lahore", "Faisalabad", 5,}
set2 = {5, 0,"Multan"}
a = set2.symmetric_difference(set1)
print(a)
```

# **Duplicate Values in Dictionary:**
In case of dictionary values, duplicates are allowed. But duplication of keys is not allowed.

*  If we use same key two times in a dictionary, the value against that key gets updated and the former value get discarded. And the dictionary will contain only one key of that name.

*  Key is a unique identifier that is used to map a specific value present against it in the dictionary. It cannot be duplicated. You cannot use  same key to map two different values.





```python
dict2 = {"name" : "Ali",
         "age12" : 21,
         "country" : "Pakistan",
         "city" : "Pakistan",
         "age" : 18,
         "age13"  : 20
         }
print(dict2)
```

To remove elements from a dictionary, following methods are used:

It is used to remove specified element from a dictionary.
*  **popitem()**
It removes and returns the last inserted key-value pair as a tuple.


```python
dict3 = {
    1 : "Python",
    2 : "C++",
    3 : "Java",
    4 : "Ruby",
    5 : "C",
}
# dict3.pop()
a = dict3.pop(4)
print(dict3)
print(a)
```
